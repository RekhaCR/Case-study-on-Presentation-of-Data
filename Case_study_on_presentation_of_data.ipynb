**4. Find the top 5 cars based on their sales price.**

```python
top_cars=df[["Model","Sale Price"]].sort_values(by=('Sale Price'),ascending=False)
#First we have to select the columns of interset and sort the value.
#Here it is asked to get top 5, that means we have to descend it.By default sort function
#will give results in ascending order,so here we have to put the argument ascending as False for getting the reverse.
top_cars.head(5)   # This will get the first five rows
```
